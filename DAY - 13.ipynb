# Step 1 – Validate Medallion Architecture

Our pipeline follows the Medallion Architecture:

Bronze → Raw event data  
Silver → Cleaned data + feature tables  
Gold → Model predictions and analytics tables

This ensures scalable and production-ready data pipelines.

# Step 2 – End-to-End Data Pipeline Flow

1️⃣ Raw Event Data → Bronze Layer
Table: bronze_events
Stores raw ecommerce clickstream events.

2️⃣ Data Cleaning → Silver Layer
Table: silver_user_features
Removes duplicates and prepares user-level aggregated features.

3️⃣ Feature Engineering
Features generated:
- total_events
- total_purchases
- total_spent
- avg_spent

4️⃣ Model Training
RandomForest model trained using Spark ML.

5️⃣ Batch Inference
Users scored with trained model.

6️⃣ Gold Layer
Table: gold_user_purchase_predictions
Stores predicted purchase probability per user.

# Step 3 – ML Lifecycle Integration

Feature Engineering  
Features derived from the Silver layer such as total_events, total_purchases, total_spent, and avg_spent.

Model Training  
RandomForest classifier trained using engineered features.

Evaluation  
Model performance evaluated using AUC metric.

Experiment Tracking  
MLflow used to track runs, parameters, and evaluation metrics.

Model Registry  
Best performing model stored and managed in MLflow.

# Step 4 – Model Retraining Strategy

Why retraining is needed:
• User behavior changes
• Data distribution shifts
• Model performance may degrade over time

Retraining triggers:
1. Scheduled retraining (weekly/monthly)
2. When model AUC drops below threshold
3. When new data volume increases significantly
4. When data drift is detected

Retraining process:
1. New event data ingested into Bronze layer
2. Features recomputed in Silver layer
3. Model retrained using updated dataset
4. New model logged to MLflow
5. Batch inference pipeline updated
